### 5- UNIFICATION OF DATAFRAMES

```python
df_estados = pd.merge(df_filtrado, df_concatenado, on=['Business_Name',	'Latitude',	'Longitude',	'Category'],  how='left')
```

```python
df_estados.tail(2)
```

```python
# eliminate the data of those businesses with zero distances
df_estados = df_estados.dropna(subset=['Distance'])
df_estados = df_estados.loc[df_estados['Distance'] < 3] 
```

```python
df_model = df_estados[['Business_Name', 'City', 'State', 'Category', 'Ranking', 'Distance', 'Tourism_Cat' ]]
df_model.drop_duplicates(inplace=True)
df_info = df_estados[['Business_Name', 'Address', 'City', 'State', 'Category', 'Distance', 'Text', 'Closest_Touristic_Business', 'Tourism_Cat']]
```

```python
# Get the unique categories
categories = df_model['Category'].unique()

# Count the total number of pubs and hotels in each state
pub_counts = df_model[df_model['Category'] == 'Pub'].groupby('State').size()
hotel_counts = df_model[df_model['Category'] == 'Hotel'].groupby('State').size()

# Create an empty DataFrame to store the filtered data
df_filtered = pd.DataFrame()

# Iterate over each category 
for category in categories:
    # Filter the businesses of the current category
    category_df = df_model[df_model['Category'] == category]
    
    # Iterate over each state and select businesses according to the specified rules
    for state in df_model['State'].unique():
        # Filter the businesses in the current state of the current category
        state_category_df = category_df[category_df['State'] == state]
        
        # Select businesses according to the specified rules
        if len(state_category_df) > 0:
            if category == 'Pub':
                max_count = 30 if state in pub_counts.index else 0
            elif category == 'Hotel':
                max_count = 30 if state in hotel_counts.index else 0
            elif category == 'Restaurant':
                max_count = min(100, 5 * len(state_category_df))
            else:
                max_count = len(state_category_df)
            
            selected_businesses = state_category_df.sample(n=min(max_count, len(state_category_df)), random_state=42)
            
            # Add the selected businesses to the filtered DataFrame
            df_filtered = pd.concat([df_filtered, selected_businesses])

# Select randomly 1000 records from the resulting DataFrame
random_sample = df_filtered.sample(n=1000, random_state=42).reset_index(drop=True)
```

```python
bus_per_cat_per_state = random_sample.groupby(['State', 'Category']).nunique().reset_index()
bus_per_cat_per_state = bus_per_cat_per_state.sort_values(by='State', ascending = False)
bus_per_cat_per_state = bus_per_cat_per_state.rename(columns={"Business_Name": "business_count"})
bus_per_cat_per_state.drop(columns={'City','Ranking','Distance','Tourism_Cat'},inplace=True)
bus_per_cat_per_state

```

```python
df_info = pd.merge(df_info, random_sample, on=['Business_Name', 'City', 'State', 'Category', 'Distance', 'Tourism_Cat'], how='inner')
```
