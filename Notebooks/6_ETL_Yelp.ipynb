### 4 - Unificación de la información - YELP

##### Observaciones:
- la información proporcionada en el archivo Tip tiene reseñas abreviadas de los usuarios. Como el archivo Review contiene información mas detallada de los comentarios y puntuaciones por parte de los usuarios, procedimos a utilizar la misma para unificar los archivos y no usar el archivo Tip.
- Las tablas se vinculan a través de business_id y user_id

Se crea un dataframe principal el cual va a unir la informacion de los dataframes:
Business, Review, Checkin y User


```python
dy_df = pd.merge(business, checkin, on= "business_id")
dy_df = pd.merge(dy_df, review, on="business_id" )
dy_df = pd.merge(dy_df, user, on="user_id" )
dy_df
```

Una vez tenemos todo el conjunto de informacion unido en un dataframe, hacemos las modificaciones necesarias:

```python
# Eliminar filas duplicadas
dy_df = dy_df.drop_duplicates().reset_index(drop = True)
dy_df
```

###### Necesitamos hacer un tratamiento con respecto a la información correspondiente a la columna "categories", desanidar la información que contiene y filtrar aquellos de nuestro interés ( negocios ya filtrados en datos correspondientes a Google Maps)

```python
# Convertir los datos str de la columna categories en lista real
dy_df["categories"] = dy_df["categories"].apply(lambda x: x.split(', ') if pd.notna(x) else [])

# Desanidado de datos que contiene la columna "categories"
rt_df = ut.explode_column(dy_df, "categories")

# Reemplazar valores nulos
rt_df = ut.replace_all_nulls(rt_df)

# Visualizar todas las categorias desanidadas
unique_categories = rt_df['categories'].unique()
for category in unique_categories:
    print(category)
```

Una vez desanidada la informacion, se filtran por aquellos que son de nuestro interés

```python
# Mencionamos aquellos negocios en los que vamos a trabajar
keywords = ['Museum', 'Park', 'Hotel', 'Motel', 'Hostel', 'Restaurant', 'Restaurant', 'Forest', 'Gallery', 'Mall', 'Pub', 'Zoo', 'Roller Coaster']

# Crear una máscara booleana para filtrar las filas
mask = rt_df['categories'].str.contains('|'.join(keywords))

# Aplicar la máscara para conservar solo las filas que contienen las palabras clave
filtered_dataframe = rt_df[mask]
```

De esta forma nos queda un dataframe filtrado solo con la informacion que nos interesa

```python
filtered_dataframe
```

 Ahora procedemos a anidar nuevamente las categorias usando la funcion group_column 

```python
##agrupar nuevamente las categorias y el resto de la información
filtered_dataframe = ut.group_column(filtered_dataframe, 'categories')
filtered_dataframe
```

```python
ut.data_summ(filtered_dataframe)
```

```python
Yelp = filtered_dataframe[["name_x", "address", "city", "state", "latitude", "longitude", "stars_x", "review_count", "date", "count_checkin", "user_id", "stars_y", "text", "name_y", "categories"]]
```

```python
Yelp = Yelp.rename(columns={
    "name_x": 'Business_Name',
    "address": 'Address',
    "city": 'City',
    "state": 'State',
    "latitude": 'Latitude',
    "longitude": 'Longitude',
    "stars_x": 'Ranking',
    "review_count": 'Review_Count',
    "date" : "Date",
    "count_checkin": 'Checkin_Count',
    "user_id": 'User_Id',
    "stars_y": 'Stars',
    "text": 'Text',
    "name_y": 'User_Name',
    "categories": 'Category'
})

```
