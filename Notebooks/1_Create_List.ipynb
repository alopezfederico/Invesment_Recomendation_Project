## Create List and Save Function

 Python Function Description

 `create_list_and_save(local_folder_path)`

Creates a list of folders and their corresponding file lists from a local folder path and saves the information to a Parquet file.

#### Parameters:
- `local_folder_path` (str): Local path of the folder to analyze.

#### Returns:
- `df` (pandas.DataFrame): DataFrame containing folder names and their associated file lists.

#### Example:
```python
create_list_and_save('/local/path/to/folder')


````python
# Create a list of folders and their corresponding file lists, then save to Parquet file
def create_list_and_save(local_folder_path,folderInList):
    """
    Creates a list of folders and their corresponding file lists from a local folder path,
    and saves the information to a Parquet file.

    Parameters:
    - local_folder_path (str): Local path of the folder to analyze.

    Returns:
    - df (pandas.DataFrame): DataFrame containing folder names and their associated file lists.

    Example:
    - create_list_and_save('/local/path/to/folder')
    """

    # Initialize an empty dictionary to store folder names and corresponding file lists
    dicc = {}

    # Extract the name of the last folder from the provided local path
    list_name = f"{local_folder_path.split('/')[-1]}"

    # Define the path for the destination Parquet file
    note_list = folderInList
    note_list_path = f'{os.path.join(note_list, list_name)}.parquet'

    # Initialize empty lists to store folder names and corresponding file lists
    name_list = []
    json_list = []

    # Check if the destination Parquet file already exists
    if not list_name in os.listdir(note_list):
        # Iterate through the sub-folders in the provided local folder path
        for sub_folder in os.listdir(local_folder_path):
            sub_folder_path = os.path.join(local_folder_path, sub_folder)

            # Check if the current item is a sub-folder
            if os.path.isdir(sub_folder_path):
                aux_name = sub_folder
                aux_list = os.listdir(sub_folder_path)
                name_list.append(aux_name)
                json_list.append(aux_list)
            else:
                # If the current item is not a sub-folder, treat it as part of the main folder
                aux_name = list_name
                aux_list = os.listdir(local_folder_path)
                name_list.append(aux_name)
                json_list.append(aux_list)
                # Break the loop since we have processed the main folder
                break

        # Populate the dictionary with folder names and corresponding file lists
        dicc['name'] = name_list
        dicc['files'] = json_list

        # Create a DataFrame from the dictionary
        df = pd.DataFrame(dicc)

        # Save the DataFrame to a Parquet file at the specified path
        df.to_parquet(path=note_list_path)

        # Return the DataFrame
        return df
    else:
        # If the destination Parquet file already exists, read and return its contents
        return pd.read_parquet(note_list_path)
````
