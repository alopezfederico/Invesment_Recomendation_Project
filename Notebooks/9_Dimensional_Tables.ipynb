### 3 - DATAFRAMES FOR ENTITY-RELATIONSHIP MODEL


#### 3.1 CATEGORY Dataframe

```python
category = pd.DataFrame(columns=["Category"])
category["Category"] = total_data["Category"].unique()
category["Category_Id"] = category.index + 1
order = ['Category_Id', 'Category']
Category = category[order]
Category
```

#### 3.2 CITY  and STATES Dataframe

```python
city = total_data[["City", "State"]]
city = city.drop_duplicates()
city = city.drop(city[city["City"] == ''].index)
city = city.sort_values(by='City', ascending=True)
city
```

```python
city = city.reset_index(drop = True)
city["City_Id"] = city.index + 1
order = ['City_Id', 'City', "State"]
city = city[order]
city
```

```python
states = total_data[["State"]]
states = states.drop_duplicates()
states = states.reset_index(drop = True)
states["State_Id"] = states.index + 1
order = ['State_Id', 'State']
states = states[order]
states
```

```python
fc = pd.merge(city, states, on="State", how="left")
fc['State_Id'] = fc['State_Id'].fillna(0).astype(int)
fc
```

```python
City = fc[["City_Id", "City", "State_Id"]]
City
```

#### 3.4 USER Dataframe

```python
user = total_data[["User_Id", "User_Name"]]
user = user.drop_duplicates()
User = user.reset_index(drop = True)
User['User_Id'] = User['User_Id'].astype(str)
User
```

#### 3.5 BUSINESS Dataframe

```python
business = total_data[["Business_Id", "Business_Name", "Address", "City", "State", "Category", "Latitude", "Longitude"]]
business = business.drop_duplicates()
business.head(3)
```

```python
bs = pd.merge(business, Category, on="Category", how="left")
bs = pd.merge(bs, states, on="State")
bs = bs.drop_duplicates()
bs = pd.merge(bs, City, on=["City", "State_Id"])
bs
```

```python
Business = bs[["Business_Id", "Business_Name", "Address", "Category_Id", "City_Id", "Latitude", "Longitude"]]
Business
```

```python
Business = Business.drop_duplicates(subset=['Business_Id'], keep='first').reset_index(drop = True)
Business
```

#### 3.6 REVIEW Dataframe

```python
review = total_data[["Business_Id", "Date", "User_Id", "Ranking", "Stars", "Text"]]
review ['Date'] = pd.to_datetime(review['Date']).dt.date
review = review.drop_duplicates().reset_index(drop = True)
review
```

```python
sentimental_analysis
```

```python
sentimental_analysis ['Date'] = pd.to_datetime(sentimental_analysis['Date']).dt.date
rv = pd.merge(review, sentimental_analysis, on=["Business_Id", "Date", "User_Id", "Ranking", "Stars", "Text"], how="right")
rv = rv.drop_duplicates()
rv["Review_Id"] = rv.index + 1
```

```python
Facts_Review = rv[["Review_Id", "Business_Id", "Date", "User_Id", "Ranking",  "Stars", "Text", "Sentiment_Analysis" ]]
Facts_Review['User_Id'] = Facts_Review['User_Id'].astype(str)
Facts_Review.reset_index(drop = True)
Facts_Review
```

#### 3.7 CHECKIN Dataframe

```python
checkin = total_data[["Business_Id", "Checkin_Count", "Date"]]
checkin = checkin.drop_duplicates()
checkin = checkin.reset_index(drop = True)
checkin["Checkin_Id"] = checkin.index + 1
checkin ['Date'] = pd.to_datetime(checkin['Date']).dt.date
checkin
```

```python
Facts_Checkin = checkin[["Checkin_Id", "Date", "Business_Id", "Checkin_Count"]]
Facts_Checkin
```
